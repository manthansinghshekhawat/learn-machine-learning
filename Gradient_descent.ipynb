```python
# Create X and y
# Using median_income as feature and median_house_value as target
X = df[['median_income']].values  # Features (2D array)
y = df['median_house_value'].values  # Target (1D array)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"First 5 X values:\n{X[:5]}")
print(f"First 5 y values:\n{y[:5]}")
```

```python
class GdRegressor :
    def __init__(self, learning_rate = 0.01 , epochs=1000):
        self.m=29.15
        self.b=-120
        self.learning_rate=learning_rate
        self.epochs=epochs
    def fit (self,X,y):
        for i in range(self.epochs):
            loss_slope=-2* np.sum(y-self.m*X.ravel()-self.b)
            # self.m=self.m-self.learning_rate*loss_slope
            self.b=self.b-self.learning_rate*loss_slope
            print(f"epoch {i} : m={self.m} , b={self.b}")
    # def predict(self,X):
	# 	return self.m*X+self.b
```

```python
gd=GdRegressor(0.1,10)
```

```python
gd.fit(X,y)
```
