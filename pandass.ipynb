```python
data = {
    "A": [1, 2, np.nan, 4, 5],
    "B": [1, 2, 3, 4, 5],
    "C": [1, 2, 3, np.nan, np.nan],
    "D": [1, np.nan, np.nan, np.nan, 5],
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df.dropna()
```

```python
df.dropna(thresh=3) ## how many non NULL values should be there in one ROW
```

```python
df.fillna(0) ## filling missing values
```

```python
values = {"A": 0, "B": 100, "C": 300, "D": 400}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
