```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,100)
```

```python
x
```

```python
y=x**2
```

```python
y= y[::-1]
```

```python
plt.plot(x,y)
plt.title("PLOT")
plt.xlabel("hola")
```

```python
### subplots
```

```python
plt.subplot(2, 2, 1)
plt.plot(x,x)
plt.subplot(2,2,2)## 4 plots totzl an second plot
plt.plot(x,y)
plt.subplot(2, 2, 3) # Third plot
plt.subplot(2, 2, 4)
plt.plot(y, x)
```

```python

## object oriented way
```

```python
fig = plt.figure()
axis1=fig.add_axes([0.1,0.1,.8,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1, 0.1, 0.4, 0.4])
axis2.plot(x, y)
plt.show()
```
